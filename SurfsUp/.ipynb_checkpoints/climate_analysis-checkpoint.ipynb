```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import datetime as dt
from datetime import timedelta
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()



# reflect the tables
Base.prepare(autoload_with=engine)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
#Use Inspector to see what's in Maresument table

inspector = inspect(engine)

columns = inspector.get_columns('measurement')
for column in columns:
    print(column["name"], column["type"])
```

```python
# Find the most recent date in the data set.

end_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()[0]
end_date
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 

# Calculate the date one year from the last date in data set.
start_date = dt.datetime.strptime(end_date, '%Y-%m-%d') - timedelta(days=365)

# Perform a query to retrieve the data and precipitation scores
data = session.query(Measurement.date, Measurement.prcp).order_by(Measurement.date.desc()).filter(Measurement.date>=start_date).all()

# Save the query results as a Pandas DataFrame. Explicitly set the column names
# Sort the dataframe by date
df_data = pd.DataFrame(data, columns=['date', 'precipitation']).sort_values('date').dropna().reset_index(drop=True)
df_data

# Use Pandas Plotting with Matplotlib to plot the data
# https://stackoverflow.com/questions/5735208/remove-the-legend-on-a-matplotlib-figure
plt.figure(figsize=(25, 12))
chart = df_data.plot('date', 'precipitation')
plt.xticks(rotation=45, ha='right', rotation_mode='anchor', size=8)
plt.yticks(size=9)
plt.title('Most recent 12 months of daily rainfall in Hawaii', y=1.05, size=15)
plt.xlabel('Date', size=12)
chart.xaxis.labelpad = 20
plt.ylabel('Inches of precipitation', size=12)
chart.yaxis.labelpad = 20
chart.get_legend().remove()
plt.savefig('Resources/rainfall.png', bbox_inches='tight')
plt.show()
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data

df_data.describe()
```

# Exploratory Station Analysis

```python
#Use Inspector to see what's in Station table

columns = inspector.get_columns('station')
for column in columns:
    print(column["name"], column["type"])
```

```python
# Design a query to calculate the total number of stations in the dataset
session.query(Station.id).count()
```

```python
# Design a query to find the most active stations (i.e. which stations have the most rows?)
# List the stations and their counts in descending order.

session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station).\
order_by(func.count(Measurement.station).desc()).all()
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.

#Store the most_active station ID as it comes up again later
most_active = session.query(Measurement.station).group_by(Measurement.station).\
order_by(func.count(Measurement.station).desc()).first()[0]

#Query for the summary stats
session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
filter(Measurement.station==most_active).all()
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

#Query for the end date
end_date = session.query(Measurement.date).filter(Measurement.station==most_active).\
order_by(Measurement.date.desc()).first()[0]

#Find the start date by subtracting a year
start_date = dt.datetime.strptime(end_date, '%Y-%m-%d') - timedelta(days=365)

#Fetch the filtered date
data = session.query(Measurement.date, Measurement.tobs).order_by(Measurement.date.desc()).\
filter(Measurement.station==most_active).filter(Measurement.date>=start_date).all()

#Make a dataframe
df_data = pd.DataFrame(data, columns=['date', 'temperature']).sort_values('date').dropna().reset_index(drop=True)


# https://stackoverflow.com/questions/67253174/how-to-set-space-between-the-axis-and-the-label
# https://stackoverflow.com/questions/12750355/python-matplotlib-figure-title-overlaps-axes-label-when-using-twiny
# https://docs.kanaries.net/tutorials/Matplotlib/matplotlib-savefig-cuts-off-labels

#Plot a histogram
plt.figure(figsize=(11,7))
chart = df_data['temperature'].hist(bins=12)
plt.title('Distribution of temperatures', y=1.04)
plt.ylabel('Frequency')
chart.yaxis.labelpad = 20
plt.xlabel('Temperature')
chart.xaxis.labelpad = 20
plt.savefig('Resources/temps.png', bbox_inches='tight')
plt.show()
```

# Close Session
